```python
import math
import decimal
from sklearn.datasets import fetch_california_housing
```

```python
def is_float(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def is_int(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

def is_calculatable(s):
    if is_int(s) or is_float(s):
        return True
    else:
        return False
```

## Задание 1
Написать программу, которая вычисляет площадь фигуры, параметры которой подаются на вход

```python
def triangle_area():
    a = str(input("1 Сторона"))
    b = str(input("2 Сторона"))
    c = str(input("3 Сторона"))

    massive_to_check = [a, b, c]

    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
        b_value = decimal.Decimal(b)
        c_value = decimal.Decimal(c)
        half_p = decimal.Decimal((a_value+b_value+c_value)/2)
    else:
        raise ValueSideException(a=a, b=b, c=c)
    return math.sqrt(float(half_p*(half_p-a_value)*(half_p-b_value)*(half_p-c_value)))

def rectangle_area():
    a = str(input("1 Сторона"))
    b = str(input("2 Сторона"))
    massive_to_check = [a, b]
    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
        b_value = decimal.Decimal(b)
    else:
        raise ValueSideException(a=a, b=b)
    return float(a_value*b_value)

def circle_area():
    a = str(input("Радиус"))
    massive_to_check = [a]
    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
    else:
        raise ValueSideException(a=a)
    return float(a_value**2 * decimal.Decimal(math.pi))

class ValueSideException(Exception):
    def __init__(self, type="form", **args):
        self.type = type
        self.values = [{key:value} for key, value in args.items()]
    
    def __str__(self):
        problem = []
        match self.type:
            case "form":
                for item in self.values:
                    if is_int(item[list(item.keys())[0]]) or is_float(item[list(item.keys())[0]]):
                        pass
                    else:
                        problem.append(item[list(item.keys())[0]])
            case "int":
                for item in self.values:
                    if is_int(item[list(item.keys())[0]]):
                        pass
                    else:
                        problem.append(item[list(item.keys())[0]])
        return f"Следующие значения не подошли:{" ".join(problem)}"

class AreaTypeException(Exception):
    def __init__(self, area_type):
        self.area_type = area_type
    def __str__(self):
        return f"Ожидалось другое значение, а пришло: {area_type=}"

def choose_figure():
    area_type=str(input("Введите интересующую Вас площадь фигруы"))
    result = dict()
    match area_type.lower().strip():
        case "треугольник":
            result = {f"{area_type.lower().strip()}": triangle_area()}
        case "прямоугольник":
            result = {f"{area_type.lower().strip()}": rectangle_area()}
        case "круг":
            result = {f"{area_type.lower().strip()}": circle_area()}
        case _:
            raise AreaTypeException(area_type=area_type)
    print(result)
```

## Задание 3
Напишите программу, которая считывает с консоли числа (по одному в строке) до тех пор, пока сумма введённых чисел не будет равна 0 и после этого выводит сумму квадратов всех считанных чисел.

```python
if __name__ == "__main__":
    float_values = []
    new_value: float = 1.0
    decimal_result = decimal.Decimal(0)
    while new_value != 0.0:
        input_value = str(input("Введите значение"))
        if is_calculatable (input_value):
            new_value = float(input_value)
            float_values.append(new_value)
        else:
            pass
    decimal_sqr_values = [decimal.Decimal(value)**2 for value in float_values]
    for decimal_value in decimal_sqr_values:
        decimal_result += decimal_value
    result = float(decimal_result)
    print({"result":result})
```

## Задание 1*
Дан текст на английском языке. Необходимо закодировать его с помощью азбуки Морзе, где каждой букве соответствует последовательность точек и тире.

На входе: В одной строке вам дан текст, который состоит из латинских
букв и пробелов.

На выходе:
Выведите каждое слово исходного текста, закодированное азбукой
Морзе. Количество строк в ответе должно совпадать с количеством слов
в исходном тексте. Между закодированными буквами ставится ровно
один пробел. Например, слово «Help» превратится в «.... . .-.. .--.».
Строчные и заглавные буквы кодируются одинаково.

```python
morze = {'a': '.-', 'b': '-...', 'c': '-.-.', 'd': '-..',
 'e': '.', 'f': '..-.', 'g': '--.', 'h': '....',
 'i': '..', 'j': '.---', 'k': '-.-', 'l': '.-..',
 'm': '--', 'n': '-.', 'o': '---', 'p': '.--.',
 'q': '--.-', 'r': '.-.', 's': '...', 't': '-',
 'u': '..-', 'v': '...-', 'w': '.--', 'x': '-..-',
 'y': '-.--', 'z': '--..'}

def translate_to_morze(some_string:str)->str:
    string_masive = some_string.split()
    new_massive = []

    for string_value in string_masive:
        for char in string_value:
            match char:
                case ".":
                    pass
                case "-":
                    pass
                case _:
                    string_value = string_value.replace(char, f"{morze[char]} ")
        new_massive.append(string_value.strip())
    result = " \n ".join(new_massive)
    return result

if __name__ == "__main__":
    print(translate_to_morze("solovyev ivan"))
```

В некотором городе открывается новая служба по доставке электронных писем. Необходимо наладить систему регистрации новых пользователей. Регистрация должна работать следующим образом: если новый
пользователь хочет зарегистрироваться на сайте, то он должен послать системе запрос name со своим именем. Система должна определить, существует ли уже такое имя в базе данных. Если такого имени не существует, то оно заносится в базу данных системы и пользователю возвращается ответ "ОК", подтверждающий успешную регистрацию. А если пользователь с таким именем уже существует, то система должна сформировать новое имя и выдать его пользователю в качестве подсказки, при этом сама подсказка также добавляется в базу данных. Новое имя формируется следующим образом: к name  оследовательно приписываются числа, начиная с 1 (name1, name2 и так далее), и среди них находят такое наименьшее i, что namei еще не содержится в системе. 

```python
def mail_system(some_requests:list):
    request_set_list_obj = {f"{i}":0 for i in list(set(some_requests))}
    new_list = []
    for request in some_requests:
        if request_set_list_obj[request] == 0:
            new_list.append("OK")
        else:
            new_list.append(f"{request}{request_set_list_obj[request]}")
        request_set_list_obj[request] += 1
    return new_list



def mail_wraper (some_list:list):
    some_number, *some_requests = some_list
    result = mail_system(some_requests=some_requests)
    return result


if __name__ == "__main__":
    print(mail_wraper([3, "b", "b", "b"]))
    print(mail_wraper([10, "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp", "bhnqaptmp"]))
    print(mail_wraper([3, "fpqhfouqdldravpjttarh", "fpqhfouqdldravpjttarh", "fpqhfouqdldravpjttarh", "fpqhfouqdldravpjttarh", "fpqhfouqdldravpjttarh", "fpqhfouqdldravpjttarh", "jmvlplnrmba", "fpqhfouqdldravpjttarh", "jmvlplnrmba", "fpqhfouqdldravpjttarh"]))
```

## Задание 3*
Необходимо создать программу обработки запросов пользователей к
файловой системе компьютера. Над каждым файлом можно производить
следующие действия: запись – w ("write"), чтение – r ("read"), запуск – x
("execute").

```python
rights= {
    "w":"write",
    "r":"read",
    "x":"execute"
}

def right_chek(right_list, actions_list):
    right_object={}
    result_list=[]
    for right_string in right_list:
        file_object, *right_file_list = right_string.split()
        right_object[f"{file_object}"] = [rights[right] for right in right_file_list]
    for action_string in actions_list:
        action_right, action_file = action_string.split()
        if action_right in right_object[action_file]:
            result_list.append("OK")
        else:
            result_list.append("Access denied")
    return result_list



def right_wrapper(some_list:list):
    right_list = some_list[1:some_list[0]+1]
    actions_list = some_list[some_list[some_list[0]+1]:]
    return right_chek(right_list, actions_list)


if __name__ == "__main__":
    print(right_wrapper([3, "python.exe x", "book.txt r w", "notebook.exe r w x", 5, "read python.exe", "read book.txt", "write notebook.exe", "execute notebook.exe", "write book.txt"]))
```
