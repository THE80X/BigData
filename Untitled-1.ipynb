```python
import math
import decimal
from sklearn.datasets import fetch_california_housing
```

```python
def is_float(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def is_int(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

def is_calculatable(s):
    if is_int(s) or is_float(s):
        return True
    else:
        return False
```

## Задание 1
Написать программу, которая вычисляет площадь фигуры, параметры которой подаются на вход

```python
def triangle_area():
    a = str(input("1 Сторона"))
    b = str(input("2 Сторона"))
    c = str(input("3 Сторона"))

    massive_to_check = [a, b, c]

    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
        b_value = decimal.Decimal(b)
        c_value = decimal.Decimal(c)
        half_p = decimal.Decimal((a_value+b_value+c_value)/2)
    else:
        raise ValueSideException(a=a, b=b, c=c)
    return math.sqrt(float(half_p*(half_p-a_value)*(half_p-b_value)*(half_p-c_value)))

def rectangle_area():
    a = str(input("1 Сторона"))
    b = str(input("2 Сторона"))
    massive_to_check = [a, b]
    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
        b_value = decimal.Decimal(b)
    else:
        raise ValueSideException(a=a, b=b)
    return float(a_value*b_value)

def circle_area():
    a = str(input("Радиус"))
    massive_to_check = [a]
    is_math = [True if is_calculatable(item_check) else False for item_check in massive_to_check]
    if False not in is_math:
        a_value = decimal.Decimal(a)
    else:
        raise ValueSideException(a=a)
    return float(a_value**2 * decimal.Decimal(math.pi))

class ValueSideException(Exception):
    def __init__(self, type="form", **args):
        self.type = type
        self.values = [{key:value} for key, value in args.items()]
    
    def __str__(self):
        problem = []
        match self.type:
            case "form":
                for item in self.values:
                    if is_int(item[list(item.keys())[0]]) or is_float(item[list(item.keys())[0]]):
                        pass
                    else:
                        problem.append(item[list(item.keys())[0]])
            case "int":
                for item in self.values:
                    if is_int(item[list(item.keys())[0]]):
                        pass
                    else:
                        problem.append(item[list(item.keys())[0]])
        return f"Следующие значения не подошли:{" ".join(problem)}"

class AreaTypeException(Exception):
    def __init__(self, area_type):
        self.area_type = area_type
    def __str__(self):
        return f"Ожидалось другое значение, а пришло: {area_type=}"


if __name__ == "__main__":
    area_type=str(input("Введите интересующую Вас площадь фигруы"))
    result = dict()
    match area_type.lower().strip():
        case "треугольник":
            result = {f"{area_type.lower().strip()}": triangle_area()}
        case "прямоугольник":
            result = {f"{area_type.lower().strip()}": rectangle_area()}
        case "круг":
            result = {f"{area_type.lower().strip()}": circle_area()}
        case _:
            raise AreaTypeException(area_type=area_type)
    print(result)
```

## Задание 2
Написать программу, которая на вход получает два числа и операцию, которую к ним нужно применить

```python
def input_variables(varibles_amount:int):
    massive_to_check = {}
    for i in range(varibles_amount):
        a = str(input(f"{i+1} число"))
        massive_to_check[f"{i+1}"]=a
    is_math = [True if is_calculatable(item_check) else False for key, item_check in massive_to_check.items()]
    if False not in is_math:
        pass
    else:
        raise ValueSideException(massive_to_check)
    result = [decimal.Decimal(value) for key, value in massive_to_check.items()]
    return result
    
def sum(a, b):
    return float(a + b)

def minus(a, b):
    return float (a - b)

def division(a, b):
    return float (a / b)

def integer_division(a, b):
    return int(float(a//b))

def absolute(a, b):
    return [float(a) if float(a)>=0 else -float(a), float(b) if float(b)>=0 else -float(b)]

def pow(a, b):
    return float(a ** b)

if __name__ == "__main__":
    new_variables = input_variables()
    a, b = new_variables[0], new_variables[1]
    result = dict()
    area_type = input("Введите операцию")
    match area_type.lower().strip():
        case "+":
            result = {f"{area_type.lower().strip()}": sum(a, b)}
        case "-":
            result = {f"{area_type.lower().strip()}": minus(a, b)}
        case "/":
            result = {f"{area_type.lower().strip()}": division(a, b)}
        case "//":
            result = {f"{area_type.lower().strip()}": integer_division(a, b)}
        case "abs":
            result = {f"{area_type.lower().strip()}": absolute(a, b)}
        case "pow":
            result = {f"{area_type.lower().strip()}": pow(a, b)}
        case "**":
            result = {f"{area_type.lower().strip()}": pow(a, b)}
        case _:
            raise AreaTypeException(area_type=area_type)
    print(result)
```

## Задание 3
Напишите программу, которая считывает с консоли числа (по одному в строке) до тех пор, пока сумма введённых чисел не будет равна 0 и после этого выводит сумму квадратов всех считанных чисел.

```python
if __name__ == "__main__":
    float_values = []
    new_value: float = 1.0
    decimal_result = decimal.Decimal(0)
    while new_value != 0.0:
        input_value = str(input("Введите значение"))
        if is_calculatable (input_value):
            new_value = float(input_value)
            float_values.append(new_value)
        else:
            pass
    decimal_sqr_values = [decimal.Decimal(value)**2 for value in float_values]
    for decimal_value in decimal_sqr_values:
        decimal_result += decimal_value
    result = float(decimal_result)
    print({"result":result})
```

## Задание 4
Напишите программу, которая выводит последовательность чисел, длинною N, где каждое число повторяется столько раз, чему оно равно.

```python
if __name__ == "__main__":
    input_value = str(input("Введите значение"))
    if is_int (input_value):
        new_value = int(input_value)
    else:
        raise ValueSideException(type="int", new_value=new_value)
    result = []
    for j in range(1, new_value+1):
        result = [*result, *[j for i in range(1, j+1)]]
    print(*result)
```

## Задание 5
Создать словарь, в котором ключи – это содержимое списка В, а значения для ключей словаря – это сумма всех элементов списка А в соответствии с буквой, содержащийся на той же позиции в списке В.

```python
def return_sum_of_vars(list_vars:list, list_keys:list):
    list_of_setted_keys = {key:0 for key in list(set(list_keys))}
    for i in range(len(list_keys)):
        if i <= len(list_vars):
            list_of_setted_keys[list_keys[i]]+=list_vars[i]
        else:
            pass
    return list_of_setted_keys


if __name__ == "__main__":
    result = return_sum_of_vars([1, 2, 3, 4, 2, 1, 3, 4, 5, 6, 5, 4, 3, 2], 
                                ['a', 'b', 'c', 'c', 'c', 'b', 'a', 'c', 'a', 'a', 'b', 'c', 'b', 'a'])
    print(result)
```
